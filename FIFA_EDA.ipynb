```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/fifa_eda.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
#checking null values
data.isnull().sum()
```

```python
#dropping the null values
data.dropna(axis = 1, inplace = True)
```

```python
data.isnull().sum()
```

```python
data.head(10)
```

```python
data.columns
```

```python
#checking the correaltion between the columns
correlation = data[['Age','Overall', 'Potential']].corr()
```

```python
# Print the correlation matrix
print("Correlation matrix:")
print(correlation)

# Plot a heatmap to visualize the correlations
plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title("Correlation between Age, Overall, and Potential")
plt.show()
```


### Key Insights:
- **Age and Overall:** Older players generally have a higher `Overall`, but the relationship isn't very strong, suggesting that there are other factors (such as experience or position) influencing the `Overall` rating.

- **Age and Potential:** Younger players tend to have higher `Potential`, as expected. The weak negative correlation shows this is a trend but not a strict rule.

- **Overall and Potential:** There’s a strong positive relationship between these two variables, indicating that players with higher current skill levels are often rated as having higher potential as well.



```python
# Now, visualize the distribution of player ages using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data['Age'], bins=20, kde=True, color='blue', edgecolor='black')
plt.title('Distribution of Player Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Categorize players into age groups
age_bins = [15, 20, 25, 30, 35, 40, 45]
age_labels = ['Under 20', '21-25', '26-30', '31-35', '36-40', 'Above 40']
data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels)
```

```python
data['Age Group']
```

```python
# Plot Overall and Potential by Age Group
plt.figure(figsize=(12, 6))

# Overall by Age Group
plt.subplot(1, 2, 1)
sns.boxplot(x='Age Group', y='Overall', data=data, palette='Set2')
plt.title('Overall by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Overall Rating')

# Potential by Age Group
plt.subplot(1, 2, 2)
sns.boxplot(x='Age Group', y='Potential', data=data, palette='Set2')
plt.title('Potential by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Potential Rating')

plt.tight_layout()
plt.show()
```

```python
#scatter plot for potential vs age
plt.figure(figsize=(8,6))
sns.scatterplot(data = data, x= "Age", y = 'Potential', color = "orange")
plt.xlabel("Age")
plt.ylabel("Potential Rating")
plt.title("Potential VS Age")
plt.show()
```

```python
# Scatter plot for Overall vs Potential
plt.figure(figsize=(8,6))
sns.scatterplot(data = data, x = 'Overall',y='Potential' ,color = "green")
plt.xlabel("Overall Rating")
plt.ylabel("Potential Rating")
plt.title("Overall VS Potential")
plt.show()
```
